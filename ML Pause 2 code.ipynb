```python
# The next step is to find true peaks in the data, to do this we find the moving mean and moving standard deviation#
# and find those signals that are a local peak, i.e we define this as 3SD above the mean at any point#
```

```python
# we define a function to get the moving mean
def get_movingMean(arr,window_size): 
    i = 0
    # Initialize an empty list to store moving mean
    moving_averages = []

    # Loop through the array to
    #consider every window of a given size
    while i < len(arr) - window_size + 1:

        # Calculate the average of the current window
        window_average = (np.sum(arr[
          i:i+window_size]) / window_size)

        # Store the average of current
        # window in moving average list
        moving_averages.append(window_average)

        # Shift window to right by one position
        i += 1
    return(moving_averages)
```

```python
# we define a function to get the moving standard deviation using a similar code

def get_movingStd(arr, window_size):
    i = 0
    # Initialize an empty list to store moving averages
    moving_stds = []

    # Loop through the array to
    #consider every window of size 200
    while i < len(arr) - window_size + 1:

        # Calculate the std of current window
        window_std = np.std(arr[i:i+window_size])

        # Store the average of current
        # window in moving std list
        moving_stds.append(window_std)
    
        # Shift window to right by one position
        i += 1
    return(moving_stds)
```

```python
# Now, we proceed to get pause status accross the data for chromosome 1, which is a point that has 3 times the SD in #
# its locality; as our default defition of a pause #

def get_pauseStatus_1():
    pause_status_1 = []
    dif_1 = my_list - padded_moving_mean_1
    mul_1 = 3 * padded_moving_std_1
    for i in range(len(my_list)):
        if dif_1[i] > mul_1[i]:
            pause_status_1.append(1)
        else:
            pause_status_1.append(0)
    return pause_status_1
```

```python
import numpy as np
import pandas as pd

def assign_signals(dataset_names, dfs, arrays_list, chr_col='chr', start_col='start', end_col='end', signal_col='signal', target_chr='chrVIII'):
    for i, dataset_name in enumerate(dataset_names):
        # Mask chr8 data
        chr8_dataset = dfs[dataset_name][dfs[dataset_name][chr_col] == target_chr]

        # Check if chr8_dataset is empty
        if chr8_dataset.empty:
            print(f"No data found for '{target_chr}' in dataset {dataset_name}")
            continue

        # Get start, end, and signal values
        starts = chr8_dataset[start_col].values
        ends = chr8_dataset[end_col].values
        signals = chr8_dataset[signal_col].values

        # Ensure all start and end positions are within bounds
        valid_indices = (starts >= 0) & (ends <= arrays_list[i].shape[0])
        starts = starts[valid_indices]
        ends = ends[valid_indices]
        signals = signals[valid_indices]

        for start, end, signal in zip(starts, ends, signals):
            arrays_list[i][start:end] = signal
```

```python
def find_median(filtered_list_nozero):
    sorted_data = sorted(filtered_list_nozero)
    n = len(sorted_data)
    mid = n // 2
    
    if n % 2 == 0:
        # Even number of elements
        median = (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        # Odd number of elements
        median = sorted_data[mid]
    
    return median
```
